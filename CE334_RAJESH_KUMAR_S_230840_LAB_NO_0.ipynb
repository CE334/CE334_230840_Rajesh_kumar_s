```python
# Inverse
a_inv = np.linalg.inv(a) # here we will use the np.linalg which is the numpy's linear algebra module which has all the matrix methods like the .inv,.svd etc
b_inv = np.linalg.inv(b)

# Rank
a_rank = np.linalg.matrix_rank(a)
b_rank = np.linalg.matrix_rank(b)

# Condition number
a_cond = np.linalg.cond(a)
b_cond = np.linalg.cond(b)

# Singular Value Decomposition
U_a, S_a, Vt_a = np.linalg.svd(a)
U_b, S_b, Vt_b = np.linalg.svd(b)

# Norm
a_norm = np.linalg.norm(a)
b_norm = np.linalg.norm(b) # here we are using the .norm function from the linalg module

print("Inverse of a:\n", a_inv)# also remember that , inverse for a matrix exist only if the determinent is not equal to zero
print("Rank of a:", a_rank)
print("Condition number of a:", a_cond)
print("Norm of a:", a_norm)
```

```python
import numpy as np

def derivative(f, a, method="central", h=0.01):# here we are using the if-elif-else condition statement to check for very condition
    if method == "forward":
        return (f(a + h) - f(a)) / h

    elif method == "backward":
        return (f(a) - f(a - h)) / h

    elif method == "central":
        return (f(a + h) - f(a - h)) / (2 * h)

    else:
        print("Invalid method")
        return None
f1 = np.exp      # e^x
f2 = np.cos     # cos(x)

x = 0.5
print("Derivative of e^x at x = 0.5:")
print("Forward :", derivative(f1, x, method="forward"))
print("Backward:", derivative(f1, x, method="backward"))
print("Central :", derivative(f1, x))   # default central

print("\nDerivative of cos(x) at x = 0.5:")
print("Forward :", derivative(f2, x, method="forward"))
print("Backward:", derivative(f2, x, method="backward"))
print("Central :", derivative(f2, x))   # default central

```
