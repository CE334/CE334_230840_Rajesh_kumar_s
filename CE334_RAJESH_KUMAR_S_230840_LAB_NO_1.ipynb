QUESTION 1

```python
import math

a = 6378137.0                 # Semi-major axis in meters
inv_f = 298.257223563         # Inverse flattening
f = 1 / inv_f                 # Flattening factor
e_sq = 2*f - f**2             # Square of eccentricity

#Function: Convert DMS to Decimal Degrees
def dms_to_dd(degrees, minutes, seconds): #Converts Degrees, Minutes, Seconds to Decimal Degrees
    return degrees + (minutes / 60) + (seconds / 3600)

def geodetic_to_cartesian(phi_deg, lam_deg, h):# this function converts the geodetic co ordinates to cartesian co ordinates

    phi = math.radians(phi_deg) # Convert degrees to radians for math functions
    lam = math.radians(lam_deg)

    # Calculate Radius of curvature in prime vertical (N)
    N = a / math.sqrt(1 - e_sq * math.sin(phi)**2)

    # Calculate X, Y, Z
    x = (N + h) * math.cos(phi) * math.cos(lam)
    y = (N + h) * math.cos(phi) * math.sin(lam)
    z = ((1 - e_sq) * N + h) * math.sin(phi)

    return x, y, z

def cartesian_to_geodetic(x, y, z):# this is the inverse function which convert the cartesian co ordinates to geodetic ordinates.
    lam = math.atan2(y, x)     #Calculate Longitude directly

    p = math.sqrt(x**2 + y**2)#Calculate p (distance from Z-axis)
    # Initial value for Latitude (phi)
    phi = math.atan2(z / p, (1 - e_sq))

    # Initialize height and old_phi for iteration
    h = 0
    old_phi = 10 # Just a starting value to ensure loop runs

    # Iterative procedure
    # We loop until the change in latitude is very small
    while abs(phi - old_phi) > 1e-12:
        old_phi = phi

        # Calculate N using current phi
        N = a / math.sqrt(1 - e_sq * math.sin(phi)**2)

        #Calculate height
        h = (p / math.cos(phi)) - N
        """here we are not using the straight formula(like phi = z/p) because the earth is not perfectly sphere and
         the Earth is flattened, this perpendicular line does not pass through the center of the Earth (unless you are at the poles or the equator). It misses the center slightly."""
        denominator = p * (1 - (e_sq * N) / (N + h))# with this update formula we can adjust that miss
        phi = math.atan(z / denominator)

    # Convert radians back to degrees for output
    phi_deg = math.degrees(phi)
    lam_deg = math.degrees(lam)

    return phi_deg, lam_deg, h

# Testing with Kanpur Data
print(" Question 1 Output")

# Kanpur Coordinates
lat_dms = (26, 26, 59.7228)
lon_dms = (80, 19, 54.7356)
h_kanpur = 126.630

# Convert to Decimal Degrees
lat_dd = dms_to_dd(*lat_dms)
lon_dd = dms_to_dd(*lon_dms)

print(f"Kanpur Input: Lat: {lat_dd:.6f}°, Lon: {lon_dd:.6f}°, H: {h_kanpur} m")

# Transform to Cartesian
x, y, z = geodetic_to_cartesian(lat_dd, lon_dd, h_kanpur)
print(f"Computed Cartesian: X: {x:.3f} m, Y: {y:.3f} m, Z: {z:.3f} m")

# Transform back to Geodetic
lat_out, lon_out, h_out = cartesian_to_geodetic(x, y, z)
print(f"Inverse Check: Lat: {lat_out:.6f}°, Lon: {lon_out:.6f}°, H: {h_out:.3f} m")

# Hometown Inputs
lat_dd = dms_to_dd(9, 10, 28)
lon_dd = dms_to_dd(77, 52, 6)
h = 106

x, y, z = geodetic_to_cartesian(lat_dd, lon_dd, h)

print(f"Hometown Cartesian Coordinates ")
print(f"Input: {lat_dd:.4f}° N, {lon_dd:.4f}° E")
print(f"X: {x:.3f} m")
print(f"Y: {y:.3f} m")
print(f"Z: {z:.3f} m")
```

```python
import math


def dms_to_dd(degrees, minutes, seconds):# this function converts the degree min sec to decimal degrees
    return degrees + (minutes / 60) + (seconds / 3600)

def spherical_to_cartesian(r, phi_deg, lam_deg):
    """
    Converts Spherical (radius, lat, lon) to Cartesian (x, y, z).
    """
    # Convert degrees to radians
    phi = math.radians(phi_deg)
    lam = math.radians(lam_deg)

    # Apply formulas
    x = r * math.cos(phi) * math.cos(lam)
    y = r * math.cos(phi) * math.sin(lam)
    z = r * math.sin(phi)

    return x, y, z

def cartesian_to_spherical(x, y, z):
    """
    Converts Cartesian (x, y, z) to Spherical (radius, lat, lon).
    """
    # 1. Calculate Radius (r)
    r = math.sqrt(x**2 + y**2 + z**2)

    # 2. Calculate Latitude (phi)
    phi = math.degrees(math.asin(z / r))# math.asin gives radians, so we convert to degrees
    # 3. Calculate Longitude
    lam = math.degrees(math.atan2(y, x))
    return r, phi, lam

print("Question 2 Output")
kanpur_lat_dms = (26, 26, 59.7228)# this is the input
kanpur_lon_dms = (80, 19, 54.7356)

# Convert to Decimal Degrees
lat_input = dms_to_dd(*kanpur_lat_dms)
lon_input = dms_to_dd(*kanpur_lon_dms)

r_input = 6378137.0 # Radius
print(f"Kanpur Spherical Input:")
print(f"Radius : {r_input} m")
print(f"Latitude  : {lat_input:.6f}°")
print(f"Longitude : {lon_input:.6f}°")
#  Transform to Cartesian
x, y, z = spherical_to_cartesian(r_input, lat_input, lon_input)
print(f"\nComputed Cartesian Coordinates:")
print(f"X: {x:.3f} m")
print(f"Y: {y:.3f} m")
print(f"Z: {z:.3f} m")
#  Transform back to Spherical
r_out, lat_out, lon_out = cartesian_to_spherical(x, y, z)
print(f"\nInverse Check:")
print(f"R: {r_out:.3f} m")
print(f"Lat: {lat_out:.6f}°")
print(f"Lon: {lon_out:.6f}°")
```

```python
import math

a = 6378137.0                 # Semi-major axis
inv_f = 298.257223563         # Inverse flattening
f = 1 / inv_f                 # Flattening factor
e_sq = 2*f - f**2             # Square of eccentricity
# here in this problem we can't convert the geodetic to spherical, so we are going through this 2 step process
def dms_to_dd(degrees, minutes, seconds):
    return degrees + (minutes / 60) + (seconds / 3600)

def geodetic_to_cartesian(phi_deg, lam_deg, h): # the function to convert the geodetic to cartesian co ordiante
    phi = math.radians(phi_deg)
    lam = math.radians(lam_deg)
    N = a / math.sqrt(1 - e_sq * math.sin(phi)**2)

    x = (N + h) * math.cos(phi) * math.cos(lam)
    y = (N + h) * math.cos(phi) * math.sin(lam)
    z = ((1 - e_sq) * N + h) * math.sin(phi)
    return x, y, z

def cartesian_to_spherical(x, y, z): # this is the function to convert the cartesian to spherical co ordinates

    r = math.sqrt(x**2 + y**2 + z**2)
    phi = math.degrees(math.asin(z / r))
    lam = math.degrees(math.atan2(y, x))
    return r, phi, lam

def geodetic_to_spherical(phi_geo, lam_geo, h_geo):
    """
    Combines the two steps to go from Geodetic directly to Spherical.
    Input: Geodetic Lat (deg), Lon (deg), Height (m)
    Output: Radius (m), Spherical Lat (deg), Spherical Lon (deg)
    """
    # Step 1: Get Cartesian coordinates
    x, y, z = geodetic_to_cartesian(phi_geo, lam_geo, h_geo)
    r, phi_sph, lam_sph = cartesian_to_spherical(x, y, z)# Step 2: Get Spherical coordinates from those Cartesian ones

    return r, phi_sph, lam_sph
#Kanpur Data
lat_geo = dms_to_dd(26, 26, 59.7228)
lon_geo = dms_to_dd(80, 19, 54.7356)
h_geo = 126.630

print(f"Geodetic Input (Kanpur):")
print(f"Lat: {lat_geo:.6f}°, Lon: {lon_geo:.6f}°, H: {h_geo} m")

# Run the transformation
r_out, lat_sph, lon_sph = geodetic_to_spherical(lat_geo, lon_geo, h_geo)

print(f"\nComputed Spherical Coordinates:")
print(f"Radius (r)     : {r_out:.3f} m")
print(f"Spherical Lat  : {lat_sph:.6f}°")
print(f"Spherical Lon  : {lon_sph:.6f}°")
```

```python
import math
import numpy as np # Using numpy makes matrix math much easier

a = 6378137.0
inv_f = 298.257223563
f = 1 / inv_f
e_sq = 2*f - f**2

def dms_to_dd(degrees, minutes, seconds):
    return degrees + (minutes / 60) + (seconds / 3600)

def geodetic_to_cartesian(phi_deg, lam_deg, h):
    phi = math.radians(phi_deg)
    lam = math.radians(lam_deg)
    N = a / math.sqrt(1 - e_sq * math.sin(phi)**2)
    x = (N + h) * math.cos(phi) * math.cos(lam)
    y = (N + h) * math.cos(phi) * math.sin(lam)
    z = ((1 - e_sq) * N + h) * math.sin(phi)
    return x, y, z

# Function: Geocentric to Topocentric (ENU) ---
def ecef_to_enu(dx, dy, dz, obs_phi_deg, obs_lam_deg): # obs_phi_deg: observer's phi, obs_lam_deg:observer's lambda
    """
    Transforms coordinate difference (dx, dy, dz) into ENU (East, North, Up).
    Uses the observer's latitude and longitude for the rotation matrix.
    """
    phi = math.radians(obs_phi_deg)
    lam = math.radians(obs_lam_deg)
    # Row 1 (East)
    r11, r12, r13 = -math.sin(lam), math.cos(lam), 0
    # Row 2 (North)
    r21, r22, r23 = -math.sin(phi)*math.cos(lam), -math.sin(phi)*math.sin(lam), math.cos(phi)
    # Row 3 (Up)
    r31, r32, r33 = math.cos(phi)*math.cos(lam), math.cos(phi)*math.sin(lam), math.sin(phi)

    # Apply Matrix Multiplication
    e = r11*dx + r12*dy + r13*dz
    n = r21*dx + r22*dy + r23*dz
    u = r31*dx + r32*dy + r33*dz
    return e, n, u

print(" Question 4 Output")
# Observer (Hometown)
obs_lat = dms_to_dd(9, 10, 28)
obs_lon = dms_to_dd(77, 52, 6)
obs_h = 106.0

# Satellite
sat_lat = dms_to_dd(26, 26, 59.7228)  # Same lat as Kanpur
sat_lon = dms_to_dd(80, 19, 54.7356)  # Same lon as Kanpur
h_kanpur_surface = 126.630
sat_altitude_from_surface = 800000.0
sat_h = h_kanpur_surface + sat_altitude_from_surface  #since the satellite is above the kanpur.

print(f"Observer (Hometown): {obs_lat:.4f}°N, {obs_lon:.4f}°E")
print(f"Satellite (Kanpur) : {sat_lat:.4f}°N, {sat_lon:.4f}°E, Altitude: 800 km")

# Convert both to Cartesian
x_obs, y_obs, z_obs = geodetic_to_cartesian(obs_lat, obs_lon, obs_h)
x_sat, y_sat, z_sat = geodetic_to_cartesian(sat_lat, sat_lon, sat_h)

#Calculate Difference Vector
dx = x_sat - x_obs
dy = y_sat - y_obs
dz = z_sat - z_obs

#Convert to Topocentric
e, n, u = ecef_to_enu(dx, dy, dz, obs_lat, obs_lon)
print(f"\nTopocentric Coordinates (Vector from Hometown to Satellite):")
print(f"East (e) : {e:.3f} m")
print(f"North (n): {n:.3f} m")
print(f"Up (u)   : {u:.3f} m")
```

```python
import math

a = 6378137.0
e_sq = 6.69437999014e-3

def dms_to_dd(d, m, s): return d + m/60 + s/3600

def geodetic_to_cartesian(phi_deg, lam_deg, h):
    phi, lam = math.radians(phi_deg), math.radians(lam_deg)
    N = a / math.sqrt(1 - e_sq * math.sin(phi)**2)
    x = (N + h) * math.cos(phi) * math.cos(lam)
    y = (N + h) * math.cos(phi) * math.sin(lam)
    z = ((1 - e_sq) * N + h) * math.sin(phi)
    return x, y, z

def cartesian_to_geodetic(x, y, z):#this is the function that converts cartesian to geodetic
    lam = math.atan2(y, x)
    p = math.sqrt(x**2 + y**2)
    phi = math.atan2(z, p * (1 - e_sq))
    h, old_phi = 0, -90
    while abs(phi - old_phi) > 1e-12:
        old_phi = phi
        N = a / math.sqrt(1 - e_sq * math.sin(phi)**2)
        h = (p / math.cos(phi)) - N
        phi = math.atan(z / (p * (1 - (e_sq * N) / (N + h))))
    return math.degrees(phi), math.degrees(lam), h

#  this is a Function to convert the Topocentric (ENU) to Geocentric (ECEF)
# Inverse of Question 4, using Transpose Matrix
def enu_to_ecef(e, n, u, obs_phi, obs_lam, x_obs, y_obs, z_obs):
    phi, lam = math.radians(obs_phi), math.radians(obs_lam)

    # Rotation Matrix Elements
    sin_p, cos_p = math.sin(phi), math.cos(phi)
    sin_l, cos_l = math.sin(lam), math.cos(lam)

    # Matrix Transpose Multiplication (Inverse Rotation)
    # dx = r11*e + r21*n + r31*u
    dx = (-sin_l * e) + (-sin_p * cos_l * n) + (cos_p * cos_l * u)
    dy = ( cos_l * e) + (-sin_p * sin_l * n) + (cos_p * sin_l * u)
    dz = (     0 * e) + ( cos_p * n)         + (sin_p * u)

    # Add difference to Observer's position
    return x_obs + dx, y_obs + dy, z_obs + dz
print(" Bonus Question Output")

obs_lat = dms_to_dd(9, 10, 28)
obs_lon = dms_to_dd(77, 52, 6)
obs_h = 106.0

target_h = 6000.0   # Aircraft is at 6km
known_e = -800.0    # East
known_n = 900.0     # North

# Observer Cartesian Position
x_obs, y_obs, z_obs = geodetic_to_cartesian(obs_lat, obs_lon, obs_h)

# Solver Loop to find correct 'u'
# Initial Guess: Aircraft Height - Observer Height
u_guess = target_h - obs_h

print(f"Goal: Find Geodetic Coords for Height = {target_h} m")
print(f"Starting Guess for u: {u_guess} m")

for i in range(10): # Usually converges in 2-3 steps
    # 1. Convert Current ENU to Cartesian
    x, y, z = enu_to_ecef(known_e, known_n, u_guess, obs_lat, obs_lon, x_obs, y_obs, z_obs)
    lat_calc, lon_calc, h_calc = cartesian_to_geodetic(x, y, z) # 2. Convert Cartesian to Geodetic to check Height
    error = target_h - h_calc    # 3. Calculate Error
    u_guess += error  # 4. Update u
    if abs(error) < 0.001: # Precision threshold
        break

print(f"\nConverged after {i+1} iterations.")
print(f"Final Topocentric Up (u): {u_guess:.3f} m")
print(f"\nFinal Aircraft Geodetic Coordinates")
print(f"Latitude : {lat_calc:.6f}°")
print(f"Longitude: {lon_calc:.6f}°")
print(f"Height   : {h_calc:.3f} m")
```
